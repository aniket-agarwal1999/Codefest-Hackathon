```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_dataset = pd.read_csv('train.csv')
test_dataset = pd.read_csv('test.csv')
```

```python
X = train_dataset.iloc[:, [1,2,3,5]].values
y = train_dataset.iloc[:, 6].values

#Encoding the category of the questions column
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
labelencoder = LabelEncoder()
X[:, 0] = labelencoder.fit_transform(X[:, 0])
onehotencoder = OneHotEncoder(categorical_features=[0])
X = onehotencoder.fit_transform(X).toarray()

#So as to remove the dependant variable paradox
X = X[:, 1:]

from sklearn.cross_validation import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.25, random_state = 0)

y_train = y_train.reshape(y_train.shape[0], 1)
y_valid = y_valid.reshape(y_valid.shape[0], 1)   #These two statements are just so as to convert them into arrays
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_valid = sc.transform(X_valid)
```

**Okay so from here on I will be using an ANN for training**

```python
###In this block I will execute neural nets so as to do the task
X_train_for_NN = X_train.T
X_valid_for_NN = X_valid.T
y_train_for_NN = y_train.T
y_valid_for_NN = y_valid.T    #taking into consideration the Andrew Ng course

#I am going to use ReLu actiavtion layers in the hidden layer's activation
def relu(x):
    return np.maximum(x,0)
    
def derivative_relu(x):
    x[x<=0] = 0
    x[x>0] = 1
    return x

#The Ann would be 2 layers deep and the units in the hidden layer would be 6
#Now I am going to use He-et-al initialization in this case

hidden_layer_units = 6
output_layer_unit = 1
w1=np.random.randn(hidden_layer_units, X_train_for_NN.shape[0])*np.sqrt(2/X_train_for_NN.shape[0])
w2=np.random.randn(output_layer_unit, hidden_layer_units)*np.sqrt(2/hidden_layer_units)
b1 = np.zeros([hidden_layer_units, 1])
b2 = np.zeros([output_layer_unit, 1])


alpha = 0.1    #learning rate
epochs = 10
for i in range(0, epochs):
    z1 = np.dot(w1,X_train_for_NN) + b1
    g1 = relu(z1)
    z2 = np.dot(w2,g1) + b2   #This would be the final output
    ##Now backpropagation will be applied
    dz2 = z2 - y_train_for_NN
    dw2 = np.dot(dz2 , g1.T)
    db2 = dz2
    dz1 = (np.dot(w2.T, dz2) * derivative_relu(z1))
    dw1 = np.dot(dz1, X_train_for_NN.T)
    db1 = dz1
    
    w2 = w2 - (alpha)*dw2
    b2 = b2 - (alpha)*db2
    w1 = w1 - (alpha)*dw1
    b1 = b1 - (alpha)*db1



```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

#initializing the ANN, that is, defining it as a sequence of layers, so we just have to create an object of sequential class here
regressor = Sequential()      #This is the future ANN that we are gonna build

#Adding the first input layer and add the first hidden layer
#Here the output_dim is equal to the number of units in the hidden layer
regressor.add(Dense(output_dim=6, init='uniform', activation='relu', input_dim=12))    #Dense basically give us everything we can do in this case
```

```python
"""
Adding the second hidden layer
regressor.add(Dense(output_dim=6, init='uniform', activation='relu'))   #In this we don't have to include the input_dim parameter   
"""
```

```python
#Adding the output layer
regressor.add(Dense(output_dim=1, init='uniform', activation=None))  #No use of input_dim
```

```python
#Compiling the ANN
regressor.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
#Fitting the ANN to our training set
regressor.fit(X_train, y_train, batch_size=10, nb_epoch=100)
```

```python
y_valid_pred = regressor.predict(X_valid)
```
